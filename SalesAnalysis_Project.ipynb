```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

data = {
    "Order_ID": range(1, 101),
    "Product": np.random.choice(["Laptop", "Phone", "Tablet", "Headphones"], 100),
    "Region": np.random.choice(["North", "South", "East", "West"], 100),
    "Sales": np.random.randint(5000, 50000, 100),
    "Quantity": np.random.randint(1, 10, 100),
    "Date": pd.date_range(start="2024-01-01", periods=100, freq='D')
}

df = pd.DataFrame(data)
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df["Month"] = df["Date"].dt.month
df["Revenue"] = df["Sales"] * df["Quantity"]
df.head()
```

```python
total_revenue = df["Revenue"].sum()
print("Total Revenue:", total_revenue)
```

```python
product_sales = df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
print(product_sales)
```

```python
region_sales = df.groupby("Region")["Revenue"].sum()
print(region_sales)
```

```python
monthly_sales = df.groupby("Month")["Revenue"].sum()
print(monthly_sales)
```

```python
product_sales.plot(kind='bar', title="Revenue by Product")
plt.ylabel("Revenue")
plt.show()
```

```python
region_sales.plot(kind='pie', autopct='%1.1f%%', title="Sales by Region")
plt.show()
```

```python
monthly_sales.plot(kind='line', marker='o', title="Monthly Revenue Trend")
plt.ylabel("Revenue")
plt.show()
```

```python
print("Business Insights:")
print("- Highest revenue product:", product_sales.idxmax())
print("- Best performing region:", region_sales.idxmax())
print("- Peak sales month:", monthly_sales.idxmax())
```
